```python
import matplotlib.pyplot as plt
import seaborn as sns

# fungsi untuk plot distribusi data
def plot_distribution(data, x_axis, y_axis, hue_col=None, title=''):
    plt.figure(figsize=(8, 5))
    c1 = sns.barplot(x=x_axis, y=y_axis, data=data,
                     palette='viridis', hue=hue_col,
                     legend=False)
    for container in c1.containers:
        c1.bar_label(container)
    plt.title(title)
    plt.show()
```

```python
def with_per(total, ax):
    for container in ax.containers:
        # mengambil semua bar (rects) dalam plot
        rects = container.get_children()
        for rect in rects:
            height = rect.get_height()
            percentage = '{:.1f}%'.format(100 * height / total)

            # menentukan posisi untuk label persentase
            x = rect.get_x() + rect.get_width() / 2
            y = height

            # menambahkan anotasi di atas setiap bar
            ax.annotate(percentage, (x, y), ha='center', va='center',
                        xytext=(0, 8), textcoords='offset points', fontsize=12)
```

```python
# mengubah tipe data kategorikal menjadi representasi numerik
from sklearn.preprocessing import LabelEncoder

data_fe = data.copy()

object_cols = [col for col in data_fe.columns if data_fe[col].dtype == 'object']

label_encoder = LabelEncoder()
for col in object_cols:
    data_fe[col] = label_encoder.fit_transform(data_fe[col])

data_fe.tail()
```

```python
# memisahkan fitur dengan label
X = data_fe.drop(columns='diagnosis')
y = data_fe['diagnosis']
```

```python
# implemetasi Chi2 sebagai seleksi fitur
from sklearn.feature_selection import chi2

# Ensure X has only non-negative values
X_non_negative = X.abs()  # Take the absolute value of all elements in X

chi_square = chi2(X_non_negative, y)

# menampilkan Chi-Score
chi_score = pd.DataFrame({'Feature': X.columns, 'Chi_Score': chi_square[0]})
chi_score = chi_score.sort_values(by='Chi_Score', ascending=False)
chi_score
```

```python
# visualisasi Chi Score
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(chi_score['Feature'], chi_score['Chi_Score'])
plt.xlabel('Features')
plt.ylabel('Chi-Square Score')
plt.title('Nilai Chi Score pada Setiap Fitur')
plt.xticks(rotation=90)
plt.show()
```

```python
# menampilkan P-Values
p_values = pd.DataFrame({'Feature': X.columns, 'P-Values': chi_square[1]})
p_values = p_values.sort_values(by='P-Values', ascending=True)
p_values
```

```python
# visualisasi p-value
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(p_values['Feature'], p_values['P-Values'])
plt.xlabel('Features')
plt.ylabel('Chi-Square P-Values')
plt.title('Nilai P-Values pada Setiap Fitur')
plt.xticks(rotation=90)
plt.show()
```

```python
# ambil fitur penting/memiliki nilai p-value < 0.05
selected_feature = p_values[p_values['P-Values'] < 0.00000000000000000005]
selected_feature
```

```python
# seleksi fitur pada dataset
X_fs = X[selected_feature['Feature'].tolist()]
X_fs
```

```python
# normalisasi dengan Min Max Normalization
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0, 1))
X_fs_sc = sc.fit_transform(X_fs)
```

# **Pemodelan**

```python
# pembagian data dengan Hold-Out Validation
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_fs_sc, y, test_size=0.3, random_state=42)
```

```python
# import matrik evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
# implementasi algoritma RF
from sklearn.naive_bayes import GaussianNB


param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}

# inisialisasi model RF
model_nbc = GaussianNB()

# inisialisasi GridSearchCV
grid_search = GridSearchCV(model_nbc, param_grid, cv=5, scoring='accuracy')

# melatih model dengan Grid Search
grid_search.fit(X_train, y_train)

# mendapatkan parameter terbaik
best_params = grid_search.best_params_

# mendapatkan model terbaik
best_model = grid_search.best_estimator_

print(best_params)
```

```python
# implementasi svm dengan parameter terbaik
best_model.fit(X_train, y_train)

y_pred_svm = best_model.predict(X_test)
print(classification_report(y_test, y_pred_svm))
```

```python
# membuat fungsi untuk menampilkan confusion matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", linewidths=0.5, linecolor="black")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()
```

```python
# menampilkan confusion matrix untuk model SVM
plot_confusion_matrix(y_test, y_pred_svm, "Confusion Matrix (SVM)")
```
